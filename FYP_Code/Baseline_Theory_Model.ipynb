**Set up the game by adding the payoff matrix values**

```python
# compute the payoff matrix
payoff_matrix = np.array([
    [1.397, 1.338, 1.368],
    [1.498, 1.412, 1.326]
    ])

print("Payoff Matrix for Offensive players:")
print(payoff_matrix)
print()
```

**Determine the best offensive action for each defensive action and vice versa**

```python
# check each column for best offensive action for each defensive action
print("Best offensive action for defensive action")
for i, defence in enumerate(["Hard-Close Out", "Moderate contest", "Sag Off"]):
  best_offensive = np.argmax(payoff_matrix[:, i])
  payoff = np.max(payoff_matrix[:, i])
  print(f"{defence}: Action {best_offensive} = {payoff}")
  print()

# check each row for best defensive action for each offensive action
print("Best defensive action for each offensive action:")
for i, offence in enumerate(["Shoot", "Drive"]):
  best_defensive = np.argmin(payoff_matrix[i, :])
  payoff = np.min(payoff_matrix[i, :])
  print(f"{offence}: Action {best_defensive} = {payoff}")
  print()
```

**Solve for Nash Equilibrium**

```python
# mixed strategy equilibirum

# create game
game = nash.Game(payoff_matrix, -payoff_matrix)

# compute equilibrium
equilibirum = list(game.support_enumeration())

# get equilibrium
offence, defence = equilibirum[0]

# offesive strategy payoffs
print("Offensive strategy payoffs:")
print(f"Shoot: {round(offence[0] * 100, 2)}%")
print(f"Drive: {round(offence[1] * 100, 2)}%")
print()

# defensive strategy payoffs
print("Defensive strategy payoffs:")
print(f"Hard-Close Out: {round(defence[0] * 100, 2)}%")
print(f"Moderate Contest: {round(defence[1] * 100, 2)}%")
print(f"Sag Off: {round(defence[2] * 100, 2)}%")
```
